# Visualization of the Orszag_Tang preblem


## import

```python
import os
import sys
from numpy import *
import matplotlib.pyplot as plt
import pyPLUTO as pypl
import pyPLUTO.pload as pp
import pyPLUTO.Image as img
```

## definition

```python
plutodir = os.environ['PLUTO_DIR']
wdir = plutodir+'/Test_Problems/MHD/Orszag_Tang/'
nlinf = pypl.nlast_info(w_dir=wdir)
#imgplot = plt.imshow(data)
```

## plot

```python
D = pp.pload(30,w_dir=wdir)
#D = pp.pload(nlinf['nlast'],w_dir=wdir)
I = img.Image()
I.pldisplay(D, D.rho,x1=D.x1,x2=D.x2,label1='x',label2='y',title=r'Density $\rho$ [Orszag Tang test]',cbar=(True,'vertical'))
plt.imshow(D.rho)
```

```python
nout = 30
for i in range(nout+1):
    D = pp.pload(i,w_dir=wdir)

    I = img.Image()
    I.pldisplay(D, D.rho,x1=D.x1,x2=D.x2,label1='x',label2='y',title=r'Density $\rho$ [Orszag Tang test]',cbar=(True,'vertical'))

    
    plt.savefig(f'vars-{i:0>4d}.png', dpi=100)
    
    plt.close(plt.gcf())
```
