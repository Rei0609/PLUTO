# Cloud Data File creation (pyFC)

```python
import pyFC
import matplotlib.pyplot as pl
import matplotlib.cm as cm
import numpy as np

pl.ion()

nc = 512
kmin = 12
fc = pyFC.LogNormalFractalCube(ni=nc, nj=nc, nk=nc, kmin=kmin, mean=1, sigma=np.sqrt(5.), beta=-5./3., )
fc.gen_cube()
```

```python
pyFC.write_cube(fc, f'input-rho_{nc}_{kmin}.flt', app=False, prec='single')
```

```python
pyFC.plot_field_stats(fc, scaling='log', vmin=-2.1, vmax=2.1, cmap=cm.jet)
```
